```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data=pd.read_excel("C:\\Users\\Dell\\Desktop\\data-analyst-data-set (1).xlsx")
```

```python
#view entire data
data
```

```python
#adding a new column for success rate
data['success_rate %']=round((data['success']*100)/data['t'],2)
data
```

```python
#data sorted on the basis of Success Rate & Hour
d1=data.sort_values(['success_rate %','hr'],ascending=True)
d1
```
